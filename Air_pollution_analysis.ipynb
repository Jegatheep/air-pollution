```python
import pandas as pd
import numpy as np
import datetime

# Define the start and end dates for the data
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)

# Generate dates for the entire year
date_range = pd.date_range(start=start_date, end=end_date)

# Create synthetic air pollution data
data = {
    'Date': date_range,
    'PM2.5': np.random.normal(loc=50, scale=10, size=len(date_range)),  # Example PM2.5 data
    'PM10': np.random.normal(loc=30, scale=8, size=len(date_range)),   # Example PM10 data
    'NO2': np.random.normal(loc=20, scale=5, size=len(date_range)),    # Example NO2 data
    'SO2': np.random.normal(loc=10, scale=3, size=len(date_range))     # Example SO2 data
}

# Create a DataFrame
air_pollution_data = pd.DataFrame(data)

# Save the data to a CSV file
air_pollution_data.to_csv('indian_air_pollution_data.csv', index=False)

# Display the first few rows of the dataset
print(air_pollution_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the air pollution data from a CSV file
data = pd.read_csv('indian_air_pollution_data.csv')

# Convert 'Date' column to datetime format and set it as index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Display the first few rows of the dataset
print(data.head())

# Basic statistics of the dataset
print(data.describe())

# Correlation matrix
correlation_matrix = data.corr()
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Air Pollution Data')
plt.show()

# Plotting time series of PM2.5 levels
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['PM2.5'], marker='o', linestyle='-', color='b')
plt.title('PM2.5 Levels Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 Level (µg/m³)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Distribution of PM2.5 levels
plt.figure(figsize=(8, 6))
sns.histplot(data['PM2.5'], kde=True, color='g')
plt.title('Distribution of PM2.5 Levels')
plt.xlabel('PM2.5 Level (µg/m³)')
plt.ylabel('Frequency')
plt.show()
```
